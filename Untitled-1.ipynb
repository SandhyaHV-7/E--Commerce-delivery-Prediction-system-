```python
# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Build deep learning model (Larger for ~700MB H5 file)
def build_model():
    model = Sequential([
        Dense(8192, activation='relu', input_shape=(X_train.shape[1],), dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(4096, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(2048, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(1024, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(512, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(256, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(128, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(64, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(32, activation='relu', dtype='float64'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(1, activation='sigmoid', dtype='float64')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Train the model (More epochs for larger file size)
model = build_model()
history = model.fit(X_train, y_train, epochs=20, batch_size=512, validation_data=(X_test, y_test))

# Save model (Ensures large file size ~700MB)
model.save("model.h5")

# Evaluate model
eval_results = model.evaluate(X_test, y_test)
print("Test Loss:", eval_results[0])
print("Test Accuracy:", eval_results[1])
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Make predictions
y_pred = (model.predict(X_test) > 0.5).astype(int)
```

```python
# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Plot confusion matrix
disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
# Plot training history (accuracy and loss)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Accuracy plot
ax[0].plot(history.history['accuracy'], label='Train Accuracy')
ax[0].plot(history.history['val_accuracy'], label='Validation Accuracy')
ax[0].set_title('Model Accuracy')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Accuracy')
ax[0].legend()
ax[0].grid()
# Loss plot
ax[1].plot(history.history['loss'], label='Train Loss')
ax[1].plot(history.history['val_loss'], label='Validation Loss')
ax[1].set_title('Model Loss')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Loss')
ax[1].legend()
ax[1].grid()

plt.show()
```
